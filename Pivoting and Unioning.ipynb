## Schools 1

```python
import pandas as pd
```

```python
schools1 = pd.read_csv(r"C:\Users\amytu\OneDrive\Documents\Capstone\Drilled Down\16_Drilled.csv")
schools1.head()
```

```python
schools1.info()
```

```python
schools1.rename(columns={'Division Name': 'Div Name','School Name': 'Sch Name', 'Pass Rate 1': '2014', 'Pass Rate 2': '2015', 'Pass Rate 3':'2015'}, inplace=True)
schools1.head()
```

```python
schools1_=schools1.drop(['2013-14 Adv Pass Rate', '2014-15 Adv Pass Rate', '2015-16 Adv Pass Rate', 'Low Grade', 'High Grade'], axis=1)
```

```python
schools1_new = pd.melt(schools1_, id_vars=["Identifier", "School Year", "Level", "Div Num", "Div Name", "Sch Num", "Sch Name", "Sch Type", "Subject", "Test"], 
                  var_name="Date", value_name="Value")
schools1_new.head()
```

```python
schools1_new.info()
```
